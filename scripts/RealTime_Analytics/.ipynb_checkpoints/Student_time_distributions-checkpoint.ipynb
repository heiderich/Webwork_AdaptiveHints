```python
import datetime
def total_time_trying(times,maxgap=5):
    """ Compute the total time in a given sequence, 
    where gaps large than maxgap minutes are set to maxgap """
    total=datetime.timedelta()
    max_time=datetime.timedelta(minutes=maxgap)
    t0=datetime.datetime.strptime(times[0],'%Y-%m-%d %H:%M:%S')
    gaps=[]
    for i in range(1,len(times)):
        t1=datetime.datetime.strptime(times[i],'%Y-%m-%d %H:%M:%S')
        gap=t1-t0
        if gap>max_time:
            gap=max_time
        gaps.append(gap.seconds/60.0);
        t0=t1
    return gaps

def time_hist_for_part(start_time,times,maxgap=30, default_start=5):
    """ Compute the time of each attempt from start_time
    If gap of first is larger than maxgap, then default_start is used."""
    max_time=datetime.timedelta(minutes=maxgap)
    t0=datetime.datetime.strptime(start_time,'%Y-%m-%d %H:%M:%S')
```
