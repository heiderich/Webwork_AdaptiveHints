```python
def parsed(string):
    try:
        expr = webwork_parser.parse_webwork(string)
        if expr:
            etree = Eval_parsed.eval_parsed(expr)
            nums = Eval_parsed.Collect_numbers(etree)
            return etree, nums
        else:
            return (None, None)
    except:
        return (None, None)        
def separate_nums(correct_etree, answer_etree):
    if not correct_etree or not answer_etree:
        return (None,None,None)
    correct_nums = frozenset(correct_etree.keys())
    answer_nums = frozenset(answer_etree.keys())
    return (frozenset(correct_nums & answer_nums),
            frozenset(answer_nums - correct_nums),
            frozenset(correct_nums - answer_nums))
```

```python
counted_incorrect_attempts = part_attempts[part_attempts['score']!='1'].groupby('answer_string').size().reset_index()
corrects=[]
incorrects=[]
missings=[]
for idx, row in counted_incorrect_attempts.iterrows():
    etree, nums = parsed(row['answer_string'])
    if etree and nums:
        correct,incorrect,missing = separate_nums(correct_nums, nums)
        corrects.append(correct)
        incorrects.append(incorrect)
        missings.append(missing)
    else:
        corrects.append(set())
        incorrects.append(set())
        missings.append(set())
```

```python
"""
This is just converting the sets into sorted strings so that we can group by them with pandas.
"""
counted_incorrect_attempts['Correct']=[str(sorted(x)) for x in corrects]
counted_incorrect_attempts['Incorrect']=[str(sorted(x)) for x in incorrects]
counted_incorrect_attempts['Missing']=[str(sorted(x)) for x in missings]
counted_incorrect_attempts.rename(columns={0: 'Count'}, inplace=True)
counted_incorrect_attempts
```

```python
""" Shows the most common sets of Correct terms, when considering all incorrect answers submitted """
counted_incorrect_attempts.groupby(['Correct']).sum().sort('Count', ascending=False)
```

```python
""" Shows the incorrect expressions, grouped by sets of correct terms """
# print counted_incorrect_attempts.groupby(['Correct', 'answer_string']).sum().to_string()
counted_incorrect_attempts.groupby(['Correct', 'answer_string']).sum()
```
