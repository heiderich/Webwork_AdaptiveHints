```python
import requests
import pandas as pd
import numpy as np

url = 'http://webwork.cse.ucsd.edu:4351/answers_by_part'
course='UCSD_CSE103'
```

```python
# course=UCSD_CSE103&set_id=Assignment10.18.13&problem_id=3
r = requests.get(url, params={'course':course})
data=r.json()
answers_by_part = pd.DataFrame(data)
answers_by_part
```

```python
answers_by_part.groupby(['set_id','problem_id','part_id']).count()
```
