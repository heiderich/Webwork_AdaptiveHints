```python
def parsed(string):
    expr = webwork_parser.parse_webwork(string)
    if expr:
        try:
            etree = Eval_parsed.eval_parsed(expr)
            nums = Eval_parsed.Collect_numbers(etree)
            return etree, nums
        except:
            return (None, None)
    else:
        return (None, None)
        
def separate_nums(correct_etree, answer_etree):
    if not correct_etree or not answer_etree:
        return (None,None,None)
    correct_nums = frozenset(correct_etree.keys())
    answer_nums = frozenset(answer_etree.keys())
    return (correct_nums & answer_nums,
            answer_nums - correct_nums,
            correct_nums - answer_nums)
```

```python
part_attempts.sort(['user_id', 'timestamp'])
correct_terms_map = defaultdict(list)
incorrect_terms_map = defaultdict(list)
missing_terms_map = defaultdict(list)
for idx, row in part_attempts.iterrows():
    if row['score']=='1':
        continue
    etree, nums = parsed(row['answer_string'])
    if etree and nums:
        correct,incorrect,missing = separate_nums(correct_nums, nums)
        correct_terms_map[correct].append(row['answer_string'])
        incorrect_terms_map[incorrect].append(row['answer_string'])
        missing_terms_map[missing].append(row['answer_string'])
```
