```python
import requests
import pandas as pd
import numpy as np
from k_medoids_clustering import ExpressionNode

url = 'http://webwork.cse.ucsd.edu:4351/answers_by_part'
course='UCSD_CSE103'
```

```python
def parse_ans(ans):
    try:
        return ExpressionNode.from_string(ans)
    except:
        return ExpressionNode((0))

def correct_dist(ans, correct_list):
    try:
        return min([ans.edit_distance(x) for x in correct_list])
    except:
        print ans

# Take the correct expression from the problem's pg file.
```

```python
r = requests.get(url, params={'course':course, 'set_id':'Assignment10.18.13', 'problem_id':1})
data=r.json()
answers_by_part = pd.DataFrame(data)
answers_by_part['answer_exps'] = answers_by_part.answer_string.map(parse_ans)
answers_by_part
```
