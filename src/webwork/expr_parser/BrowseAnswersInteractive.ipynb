```python
import requests
import pandas as pd
import numpy as np
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from IPython.display import clear_output, display, HTML
from collections import OrderedDict
```

```python
# Load all course data
url = 'http://webwork.cse.ucsd.edu:4351/answers_by_part'
course='UCSD_CSE103'
r = requests.get(url, params={'course':course})
data=r.json()
DF = pd.DataFrame(data)
```

```python
# Create dropdowns
s=DF.groupby('set_id').size()
s.sort(ascending=False)
sets = OrderedDict([("{0} - {1}".format(k,v), k) for k,v in s.iteritems()])

assignment_picker = widgets.DropdownWidget(values=sets, description="Assignment:")
problem_picker = widgets.DropdownWidget(values=[], description="Problem")
part_picker = widgets.DropdownWidget(values=[], description="Part")
user_picker = widgets.DropdownWidget(values=[], description="User")

# Defaults
assignment = sets.values()[0]
problem = 0
part = 0
user=""
data = DF

# Change listeners
def on_assignment_change(trait_name, new_value):
    global assignment
    assignment = new_value
    update_problems()
    update_users()
    refilter()
    
def on_problem_change(trait_name, new_value):
    global problem
    problem = new_value
    update_parts()
    update_users()
    refilter()

def on_part_change(trait_name, new_value):
    global part
    part = new_value
    update_users()
    refilter()
    
def on_user_change(trait_name, new_value):
    global user
    user = new_value
    refilter()

def update_problems():
    problems = DF[DF['set_id']==assignment].groupby('problem_id').size()
    problems.sort_index()
    problem_picker.values = OrderedDict([("{0} - {1}".format(k,v),k) for k,v in problems.iteritems()])
    
def update_parts():
    parts = DF[(DF['set_id']==assignment) & (DF['problem_id']==problem)].groupby('part_id').size()
    parts.sort_index()
    part_picker.values = OrderedDict([("{0} - {1}".format(k,v),k) for k,v in parts.iteritems()])
    
def update_users():
    global user
    user=""
    user_counts = DF[(DF['set_id']==assignment) & (DF['problem_id']==problem) & (DF['part_id']==part)].groupby('user_id').size()
    user_counts.sort(ascending=False)
    user_picker.values = OrderedDict([("{0} - {1}".format(k,v), k) for k,v in user_counts.iteritems()])

def refilter():
    global data
    data = DF[(DF['set_id']==assignment) & (DF['problem_id']==problem) & (DF['part_id']==part) & (DF['user_id'].str.startswith(user))]
    
assignment_picker.on_trait_change(on_assignment_change, 'value')
problem_picker.on_trait_change(on_problem_change, 'value')
part_picker.on_trait_change(on_part_change, 'value')
user_picker.on_trait_change(on_user_change, 'value')
```
