```python
hints = logs[logs['part_no'] > 1000]

def get_hint_context(logs, hint, minutes=60):
    ''' Given a row in the logs table corresponding to a hint, 
        get the attempts at the same problem by the same user
        within <minutes> minutes of the hint '''        
    # Analysis for a particular hint
    t = hint['timestamp']
    m10 = timedelta(minutes=minutes)
    # the answer log surrounding a hint
    context = logs[(logs['timestamp'] >= t - m10) & 
                   (logs['timestamp'] < t + m10)  & 
                   (logs['user'] == hint['user']) &
                   (logs['Assignment'] == hint['Assignment']) &
                   (logs['problem_no'] == hint['problem_no'])
                   ]
    return context

def evaluate_hint_effectiveness(context, hint):
    ''' Given the answer logs surrounding a hint, and the log entry for the hint of interest,
        find the part the student struggled with most just before the hint.  
        Return the time spent on this part before/after, and whether the student answered the part correctly
        after the hint.  
        Returns None if the student answered every part before the hint, in the given context, correctly
    '''
    global a,b,c
    a += 1
    # Get index of hint
    correct_hint_answers = ((context['part_no'] == hint['part_no']) & context['correct']).values
    if not any(correct_hint_answers): #The hint was never answered correctly
        b += 1
        return None
    hint_index = correct_hint_answers.tolist().index(True)
    # Get non-hint context before/after first attempt at hint
    before = context[:hint_index]
    after = context[hint_index:]
    # Remove hints from before/after hint context
    before = before[before['part_no'] <= 1000]
    after = after[after['part_no'] <= 1000]
    part_before_times = before.groupby('part_no')['timediff'].sum()
    part_before_tries = before.groupby('part_no').size()
    part_correctness = before.groupby('part_no')['correct'].max()
    # Define "incorrect time" for a log and part as the time spent on incorrect attempts at a problem
    incorrect_parts = part_correctness[~part_correctness].index
    if len(incorrect_parts) == 0: #There were no incorrect attempts at non-hints within an hour before the hint
        c += 1
        return None
    part_before_times.reindex(incorrect_parts) #Get rid of parts that were answered correctly
    part_before_tries.reindex(incorrect_parts) #Get rid of parts that were answered correctly before the hint
    # Find the part with the largest incorrect time before the hint, that was not answered correctly before the hint
    #hard_part = part_before_times.idxmax()
    # Instead of finding the hardest part, find the part the hint is associated with in the <assigned_hint> table
    # This is what we really should be looking at
    try:
        hard_part = assigned_hints.ix[hint['part_no']]['part_no']
    except KeyError: #assigned hint was probably deleted
        return None
    # Calculate the before/after incorrect times.  <after incorrect time>/<before> = hint effectiveness
    before_time = before[before['part_no'] == hard_part]['timediff'].sum()
    after_time = after[after['part_no'] == hard_part]['timediff'].sum()
    before_tries = (before['part_no'] == hard_part).sum()
    after_tries = (after['part_no'] == hard_part).sum()
    # Did the student eventually get the hard part right, after the given hint?
    after_correct = after[after['part_no'] == hard_part]['correct'].max()
    return (before_time,after_time,before_tries,after_tries,after_correct)
```

### Statistical Significance ###
We want to quantify hint effectiveness.  Throughout the course, 792 hints were assigned. We want to determine a causal effect of the hints on the number of attempts a student makes after the problem is given.  

Each hint corresponds to one answer box.  

We've prefiltered: 
* only hints that were eventually answered correctly are kept
* only problems where a student was unable to answer the box corresponding to the hint correctly before, but answered it successfully after, are kept

After this filtering, there are 440 remaining hint/boxes that were assigned.  Note, a student could make multiple attempts at each such hint/box.  

Among these cases, we want to show that the a student correctly answering a hint significantly decreases the remaining number of tries that student will make, until answering the question successfully.  

To do this, we compare the average number of tries after hints in the 440-size dataset, to the number of tries we would expect, assuming hints had no causal effect on the number of tries remaining.  

The hints system we built allows an instructor to sift through students that are struggling, and dynamically give hints.  This determines the distribution of hints after a number of student unsuccessful tries at a problem.  

Assuming that hints are randomly assigned based on the empirical distribution of the number of tries, and hints have no effect on the number of tries remaining, what is an estimation of the expected number of tries?  

We approximated this using a monte-carlo simulation.  We randomly permuted the assignment of problem boxes to hints (so pair-wise relationships are ignored), and calculated the number of tries between when the hint was assigned and when the student successfully answered the problem.  

In the alternative case, we computed the average number of tries between when the hint was assigned and when the student answered the problem correctly.  

Notice: by keeping only problems where hints were assigned in our null distribution of tries, we correct for the bias in choices of problems to assign hints.  The tries distribution for problems where hints were assigned probably differs from the overall distribution of tries.  Luckily, hints were assigned to enough problems so that we still have significance.  

Over the 440 hint/problems/users, the number of tries between when the student answered the hint correctly and when the student answered the problem correctly had average 6.157 and empirical standard deviation 9.251.  

In the case where the relationships between hints and problems were removed (the assignment of hints to problems was randomly permuted), the average number of tries between when the student answered the hint correctly and when the student answered the problem correctly had an average of 10.489 and a standard deviation of 14.46.  

Since our sample size is 440, we can safely assume normality of the sample average.  Loosely bounding the standard deviation at 15, the difference in means is very significant.  The z-score is $$\frac{10.489-6.157}{15/\sqrt{440}} \approx 6.0578$$.
