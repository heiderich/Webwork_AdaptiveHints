## Scatter Plots ##

```python
def nanVsMeanAnalysis(Table,name,axis=1):    
    s=shape(Table)
    sz=s[0]*s[1]+0.0
    count=Table.count().sum() # count the number of entries that are not nan
    print 'The shape of the table is %d X %d=%d. Out of these %d are defined, which is %f3.1 percent' % (s[0],s[1],sz,count,(0.0+sz-count)/sz)
    valid=Table.count(axis=axis)
    xlabel('no. of valid entries (not nan)')
    ylabel('mean of '+name)
    grid()
    if axis==0: per='problem'
    else: per='user'
    title(name+' vs number of defined entries, per '+per)
    meanVal=Table.mean(axis=axis)
    scatter(valid,meanVal)
```
