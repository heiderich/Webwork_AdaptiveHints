## Scatter Plots ##

```python
def nanVsMeanAnalysis(Fc,Table,name,axis=1,split=True): 
    s=shape(Table)
    sz=s[0]*s[1]+0.0
    count=Table.count().sum() # count the number of entries that are not nan
    print 'The shape of the table is %d X %d=%d. Out of these %d are defined, which is %f3.1 percent' % (s[0],s[1],sz,count,100.0*(0.0+count)/sz)
    valid=Fc.count(axis=axis)
    fig=figure(figsize=[8,8])
    xlabel('no. of valid entries (not nan)')
    ylabel('mean of '+name)
    grid()
    if axis==0: per='problem'
    else: per='user'
    title(name+' vs number of defined entries, per '+per)
    if split:
        meanValC=Table.where((Fc==1).values).mean(axis=axis)
        meanValI=Table.where((Fc==0).values).mean(axis=axis)
        print 'valid=',shape(valid),', meanValC=',shape(meanValC)
        scatter(valid,meanValC,c='b')
        scatter(valid,meanValI,c='r')
    else:
        meanVal=Table.mean(axis=axis)
        print 'valid=',shape(valid),', meanVal=',shape(meanVal)
        scatter(valid,meanVal,c='b')
```
