```python
import requests
import pandas as pd
import numpy as np
from k_medoids_clustering import ExpressionNode

url = 'http://webwork.cse.ucsd.edu:4351/answers_by_part'
course='UCSD_CSE103'
```

```python
r = requests.get(url, params={'course':course, 'set_id':'Assignment10.16.13', 'problem_id':1})
data=r.json()
answers_by_part = pd.DataFrame(data)
answers_by_part
```

```python
correct_answers = answers_by_part[answers_by_part.score=='1'].answer_string.unique()

def parse_ans(ans):
    try:
        return ExpressionNode.from_string(ans)
    except:
        return ExpressionNode((0))

def correct_dist(ans, correct_list):
    try:
        return min([ans.edit_distance(x) for x in correct_list])
    except:
        print ans
answer_exps = answers_by_part.answer_string.map(parse_ans)
correct_exps = [parse_ans(ans) for ans in correct_answers]

correct_dists = answer_exps.map(lambda ans: correct_dist(ans, correct_exps))
```

```python
answers_by_part['correct_distance'] = correct_dists
answers_by_part
```
