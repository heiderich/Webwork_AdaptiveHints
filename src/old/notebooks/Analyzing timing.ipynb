```python
def sample_df(df, samples):
    ''' Samples rows of a dataframe '''
    if len(df)<samples:
        return df
    else:
        rows = random.sample(df.index, samples)
        return df.ix[rows]
```

```python
def diff(S):
    A=S.values
    D=A[1:]-A[:-1]
    if len(D)>0:
        if min(D)<0:
            sys.stderr.write('negative time difference')
            exit(keep_kernel=True)
    D=array([d if d<600 else np.nan for d in D])
    return D
```

```python
Diffs=logs.groupby(['user','Assignment','problem_no','part_no'])['time'].apply(diff)
```

```python
PerUserPart=pd.DataFrame({'TimeList':Diffs})
PerUserPart[:6]
```

```python
PerUserPart['sum']=PerUserPart['TimeList'].apply(sum)
PerUserPart['cnt']=PerUserPart['TimeList'].apply(len)
```

```python
PerUserPart[:10]
```

```python
TriesTable=PerUserPart.unstack(level='user')
TriesTable.ix('Assignment11.04.13')
```

```python

ProblemStat=pd.DataFrame({'std Count':TriesTable['cnt'].std(1),'mean Count':TriesTable['cnt'].mean(1),'mean Time':TriesTable['sum'].mean(1),'# of not NaN':TriesTable['cnt'].count(1)})
```

```python
ProblemStat=ProblemStat[ProblemStat['# of not NaN']>50]
ProblemStat.sort(column='mean Count',ascending=False)
```
